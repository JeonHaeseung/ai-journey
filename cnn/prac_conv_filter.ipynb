```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Image by Andrew Whitmore
CURRENT_DIR = os.getcwd()
IMAGE_PATH = os.path.join(CURRENT_DIR, "..", "image", "building.png")
image = cv2.imread(IMAGE_PATH)
print(f"Image shape: {image.shape}")
```

```python
def show_image(image, result):
	plt.figure(figsize=(10,6))
	plt.subplot(121)
	plt.imshow(image)
	plt.subplot(122)
	plt.imshow(result)
	plt.show()
```

## Blur

```python
smallBlur = np.ones((7, 7), dtype=np.float32) * (1.0 / (7 * 7))
result = cv2.filter2D(image, -1, smallBlur)
show_image(image, result)
```

## Sharpen

```python
sharpen = np.array((
	[ 0, -1,  0],
	[-1,  5, -1],
	[ 0, -1,  0]), dtype="int")

result = cv2.filter2D(image, -1, sharpen)
show_image(image, result)
```

## Edge detection

```python
# Detect every edge
laplacian = np.array((
	[0, 1, 0],
	[1, -4, 1],
	[0, 1, 0]), dtype="int")

result = cv2.filter2D(image, -1, laplacian)
show_image(image, result)
```

```python
# Detact horizontal edge
sobel_h = np.array((
	[-1, 0, 1],
	[-2, 0, 2],
	[-1, 0, 1]), dtype="int")

result = cv2.filter2D(image, -1, sobel_h)
show_image(image, result)
```

```python
# Detact vertical edges
sobel_v = np.array((
	[-1, -1, -1],
	[ 2,  2,  2],
	[-1, -1, -1]), dtype="int")

result = cv2.filter2D(image, -1, sobel_v)
show_image(image, result)
```

```python
# Detact vertical edges
sobel_v = np.array((
	[-1, -2, -1],
	[ 0,  0,  0],
	[ 1,  2,  1]), dtype="int")

result = cv2.filter2D(image, -1, sobel_v)
show_image(image, result)
```
